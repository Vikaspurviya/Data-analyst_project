```python
data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'Salesperson': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Hank'],
    'Product': ['Laptop', 'Laptop', 'Tablet', 'Tablet', 'Phone', 'Phone', 'Laptop', 'Tablet'],
    'Units_Sold': [120, 150, 90, 130, 160, 100, 200, 170],
    'Revenue': [120000, 150000, 90000, 130000, 160000, 100000, 200000, 170000]
}

df = pd.DataFrame(data)
print(df)
```

```python
pd.pivot_table(df,values='Revenue',index='Region',columns='Product')
```

```python
pd.pivot_table(df,values=['Revenue','Units_Sold'],index='Region',columns='Product')
```

```python
pd.crosstab(df['Region'],df['Product'])
```
