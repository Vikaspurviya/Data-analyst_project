```python
import numpy as np
a = np.array([20,40,60,70])
print(np.sum(a))
print(np.min(a))
print(np.max(a))
print(np.size(a))
print(np.mean(a))
print(np.median(a))
print(np.cumsum(a))
print(np.cumprod(a))
```

```python
a =[100,150,199,200,250,130]
b=[10,50,30,40,30,10]
price=np.array(a)
quantity=np.array(b)
print(price,"\n",quantity)

c=(np.cumprod([price,quantity],axis = 0))
print(c)
print(c[1].sum())
```

```python
import numpy as np
import statistics as st
baked_food=[200,150,150,130,200,220,170,188] 
a = np.array(baked_food)
print(np.sort(a))
print(np.mean(a))   #sum of all the value/num of  the value
print(np.median(a))  
print(st.mode(a))
print(np.std(a))
print(np.var(a)) 
```

Cofficent of co-relation check dependence on data

```python
import numpy as np
tobacco_consumption =np.array([30,50,10,30,50,40])
deaths =np.array([100,120,70,100,120,112])
print(np.corrcoef([tobacco_consumption,deaths])) # -1 is strong relation and +1 is weak relation and 0 means neutral
```
