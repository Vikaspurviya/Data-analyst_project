```python
import matplotlib.pyplot as plt

Months = ['Jan', 'Feb', 'Mar', 'April']
Sales1 = [1400, 200, 1500, 5000]


plt.plot(Months, Sales1, color='red', linestyle='--', linewidth=2, marker='o',label="2025 sales data")
plt.xlabel("Months")
plt.ylabel("Sales (in Rs)")
plt.title("Monthly Sales Report 2025")
plt.legend(loc='upper left',fontsize=12)
plt.grid(color='gray',linestyle=':',linewidth='1')
plt.xlim(1,4)
plt.ylim(200,10000)
plt.xticks(['Jan','Feb','Mar','April'],['M1','M2','M3','M4'])
plt.show()
```

```python
import matplotlib.pyplot as plt
regions = ['north','east','west','south']
revenue=[3000,2000,4000,1000]

plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['blue','yellow','lightgreen','gold'])
plt.title('Revenue Contribution by region')

plt.show()
```

```python
import matplotlib.pyplot as plt
score = [77,75,91,81,91,67,88,54,45,78,68,48,56,76,75,89,95,55,66,78]
plt.hist(score,bins=5 ,color='red',edgecolor='black')
plt.xlabel('Score range')
plt.ylabel('No. of student')
plt.title('score distribution of student')
plt.show()
```

plt.scatter(x,y,color="color_name",marker='o',label="label_name")

```python
import matplotlib.pyplot as plt
hours_studies = [1,2,3,4,5,6,7,8]
exam_score =[50,55,60,65,70,75,80,90]
plt.scatter(hours_studies,exam_score,color='red',marker='o',label='Student data')
plt.xlabel('Hour studied')
plt.ylabel('Exam score')
plt.legend()
plt.title("Relationship between study time and exam score")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter([1,2,3],[50,60,70], color='red',marker='o',label='Class A')
plt.scatter([1,2,3],[45,50,52], color='Orange',marker='o',label='Class B')
plt.xlabel('Hour studied')
plt.ylabel('Exam score')
plt.legend()
plt.title("Camparision of two classes")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[10,20,15,25]
plt.subplot(1,2,1)
plt.plot(x,y,color='green')
plt.title("Line chart")
plt.subplot(1,2,2)
plt.bar(x,y,color='orange')
plt.title('Bar chart')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[10,20,15,25]

fig, ax = plt.subplots(1,2,figsize=(10,5))
ax[0].plot(x,y)
ax[0].set_title("Line chart")

ax[1].bar(x,y)
ax[1].set_title("Bar graph")
fig.suptitle('Camparision of line and bar chart')
plt.tight_layout()
plt.show()
```
