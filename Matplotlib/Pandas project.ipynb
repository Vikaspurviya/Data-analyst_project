```python
def extract_time(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i]==')':
            for j in range(i+1,i+20):
                data +=txt[j]
            return data
```

```python
df['Total_time']=df['Title'].apply(extract_time)
df
```

```python
# Function to calculate total months
def calculate_months(duration):
    start, end = duration.split(' - ')
    start_date = pd.to_datetime(start, format='%b %Y')
    end_date = pd.to_datetime(end, format='%b %Y')
    # Difference in months
    total_months = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month)
    return total_months

# Apply the function to the column
df['Total Months'] = df['Total_time'].apply(calculate_months)
df
```
